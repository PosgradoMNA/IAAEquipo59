```python
# obtenemos el user_ID, name 
df_rat = df_rat[['userID', 'placeID', 'rating']]
df_geo = df_geo[['placeID', 'name']]

# unimos los dos dataframes
df = pd.merge(df_rat, df_geo, on='placeID')

# imprimimos el shape del dataframe

print('Tamaño matriz de utilidad: ', df.shape)

# imprimimos los primeros 10 registros
df.head(10)
```

# **Ejercicio-2.**

```python
df.info()
```

```python
# Generamos la matriz de utilidad y reemplazamos los valores NaN por 0

UtMx = df.pivot_table(values='rating', index='userID', columns='name', fill_value=0)

UtMx.head(10)
```

```python
# obtenemos el número de usuarios y restaurantes
UtMx.shape # vemos que tenemos 138 usuarios y 129 restaurantes
```

```python
# obtenemos la proporción de los datos no ceros

print('Total de elementos de la matriz de utilidad: %d' % (UtMx.size))
print('Total de elementos diferentes de cero: %d' % (np.count_nonzero(UtMx)))
print('Porcentaje de elementos diferentes de cero: %.1f%%' % (100 * np.count_nonzero(UtMx) / UtMx.size))
```

```python
# Creamos la matrix de utilidad transpuesta 

X = UtMx.T
X.shape
```

```python
X.head(10)
```

```python
# obtenemos el numero componentes principales

n_components = 137
svd = TruncatedSVD(n_components=n_components, random_state=1)
svd.fit(X)

# obtenemos la varianza explicada por cada componente principal

var_exp = svd.explained_variance_ratio_

# obtenemos la varianza acumulada

var_exp_cum = np.cumsum(var_exp)
```

```python
# utilizando el número de componentes principales obtenido, creamos el modelo
num_sv = 137
svd = TruncatedSVD(n_components=num_sv, random_state=1)
resultant_matrix = svd.fit_transform(X)

# imprimimos el shape de la matriz resultante
resultant_matrix.shape

# creamos un dataframe con los componentes principales

df_comp = pd.DataFrame(resultant_matrix, index=X.index)
df_comp.head(5)
```

```python
# imprimos los valores singulares obtenidos

print('Valores singulares: ', resultant_matrix)
```

# **Ejercicio-3.**

```python
# Definimos la matriz de correlación entre los componentes principales

corrMx = np.corrcoef(resultant_matrix)
corrMx.shape
```
